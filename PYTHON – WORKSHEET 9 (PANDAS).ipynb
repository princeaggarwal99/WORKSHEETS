# Q.11 - Q.15 solutions

```python
import pandas as pd
import numpy as np
```

```python
# Q.11 Write a Pandas program to create and display a DataFrame from the following dictionary data and labels:

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin',
'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

# making into a dataframe :

data=pd.DataFrame(exam_data)
data.index=['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X']
data
```

```python
#Q.12  Write a Pandas program to get the first 5 rows of the DataFrame created in Q11.

data.head(5)
```

```python
# Q.13. Write a Pandas program to select the 'name' and 'score' columns of the DataFrame created in Q11.

data[['name','score']]
```

```python
# Q.14 Write a Pandas program to select 'name' and 'score' columns in row indexes 3, 5, 6, 8 from the DataFrame created in Q11.

data.loc[['III','V','VI','VIII'],['name','score']]
```

```python
# Q.15 Write a Pandas program to select the rows where the score is between 15 and 20 (inclusive) from the DataFrame created in Q11.

data[data['score'].between(15,20)]
```
